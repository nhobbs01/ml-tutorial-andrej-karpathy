Insertion Sort

```python
def insertion_sort(A):
    for i in range(1,len(A)):
        key = A[i]
        j = i-1
        #print(f'i: {i}, j: {j}, key: {key}')
        while j>=0 and A[j] > key:
            A[j+1] = A[j]
            j = j -1
        A[j+1] = key
        print(f'{A}')

    return A
```

```python
x = [4,2,5,1,3,5 ,0]
print(x)
res = insertion_sort(x)
```

```python
def insertion_sort_rev(A: list[int]) -> list[int]:
    for i in range(1, len(A)):
        key = A[i]
        j = i - 1
        while j >= 0 and key > A[j]:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = key
    return A
```

```python
insertion_sort_rev(x)
```

```python
def add_binary_integers(A, B, n):
    carry = 0
    C = [0] * n
    for i in range(n-1): 
        x = (A[i] + B[i] + carry) % 2
        if(A[i] and B[i]):
            carry = 1
        C[i] = x
    C[n-1] = carry
    return C
        
```

```python
a = [0,1,1,0]
b = [1,1,0,1]
add_binary_integers(a, b, 5)
```

```python
def selection_sort(A):
    for i in range(len(A)):
        key = A[i]
        # loop through A[i:] to find smallest
        smallest = i
        for j in range(i+1, len(A)):
            if A[j] < A[smallest]:
                smallest = j
        # Swap with A[1]
        # print(A[smallest])
        temp = A[smallest]
        A[smallest] = key
        A[i] = temp
        # print(A)
    return A
```

```python
x = [5,2,6,4, 10 ,2, -1]
selection_sort(x)
# O(n^2)
```

```python
# Merge step for merge sort
# A is the original Array
# p is the start index
# q is the midpoint
# r is the end point
def merge(A, p, q, r):
    # Left array
    L = A[p:q]
    R = A[q: r]
    i = 0
    j = 0
    k = p # place in A to replace
    while i<len(L) and j<len(R):
        if(L[i] < R[j]):
            A[k] = L[i]
            i+=1
        else:
            A[k] = R[j]
            j += 1
        k+=1
    while i < len(L):
        A[k] = L[i]
        k+=1
        i+=1
    while j < len(R):
        A[k] = R[j]
        k+=1
        i++1
    return A
```

```python
def merge_sort(A, p, r):
    # Base case (zero or 1 element)
    if(r-p <= 1):
        print('return')
        return
    q = (p+r)//2
    
    merge_sort(A,p,q)
    merge_sort(A,q,r)
    merge(A, p, q, r)
    
```

```python
a = [2,6,3,6,2,4]
merge_sort(a, 0, 4)
```
